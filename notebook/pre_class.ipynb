### 2. 基础操作与快捷键 (Cheat Sheet)

Jupyter 的操作核心是 **Cell (单元格)**。

*   **运行代码**: `Shift + Enter` (运行并选中下一行，**最常用**)
*   **运行代码**: `Ctrl + Enter` (运行并停留在当前行)
*   **切换模式**:
    *   `Esc`: 进入**命令模式** (左侧蓝条)，可以按 `A` (Above) 在上方插入，按 `B` (Below) 在下方插入。
    *   `Enter`: 进入**编辑模式** (左侧蓝条变色)，开始写代码。
*   **切换类型**:
    *   `M`: 变为 **Markdown** (写笔记)。
    *   `Y`: 变为 **Code** (写代码)。

Cell 分为 **Code Cell** 和 **Markdown Cell** 两种类型，分别用于写代码和写笔记。

Jupyter 支持 LaTeX 语法，可以在 Markdown Cell 中写数学公式，例如：

例如，著名的欧拉公式：
$$ e^{i\pi} + 1 = 0 $$

或者正态分布公式：
$$ f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{ -\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2 } $$

### 3. 输出机制：Implicit Display vs Print

Jupyter 有一个 **REPL (Read-Eval-Print Loop)** 特性：
*   **默认行为**：Cell 的**最后一行**如果是一个变量，Jupyter 会自动调用 `display()` 展示它。
*   **坑点**：如果你在一个 Cell 里写了多个变量，默认只会显示最后一个！

```python
from IPython.display import display

a = "Hello Output"
b = "Hidden Variable"
c = {"Course": "Pre-lecture", "Role": "TA", "Name": "Gu Yi"}

print("--- 1. 普通 Python 脚本的行为 ---")
a  # 哑巴：在普通脚本里这行没用，在 Jupyter 若不是最后一行也不显示

print("--- 2. Jupyter 的默认行为 (REPL) ---")
# 只有最后一行会被自动 display
c
```

```python
print("--- 3. 如果想看所有变量怎么办？ ---")
# 使用 display() 函数，它比 print() 更强大，能渲染富文本
display(a)
display(c)
```

### 5. 可视化与交互 (Interactive)

Jupyter 最方便的地方在于“所见即所得”的交互。
我们使用 `ipywidgets` (基于 IPython 的组件) 来实现。

```python
# 这一步需要安装 matplotlib 和 ipywidgets
import matplotlib.pyplot as plt
from ipywidgets import interact


# 定义一个绘图函数
def plot_wave(frequency=1.0, amplitude=1.0, color="blue"):
    x = np.linspace(0, 2 * np.pi, 200)
    y = amplitude * np.sin(frequency * x)

    plt.figure(figsize=(8, 4))
    plt.plot(x, y, color=color, linewidth=3)
    plt.ylim(-5.5, 5.5)
    plt.grid(True, alpha=0.3)
    plt.title(f"Wave: y = {amplitude} * sin({frequency} * x)")
    plt.show()


# 启动交互！拖动滑块试试
print("交互式演示：请拖动下方滑块")
interact(
    plot_wave,
    frequency=(0.5, 5.0, 0.1),
    amplitude=(0.1, 5.0, 0.1),
    color=["blue", "red", "green", "orange"],
);
```

```python
import plotly.graph_objects as go
import numpy as np

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x_grid, y_grid = np.meshgrid(x, y)
r = np.sqrt(x_grid**2 + y_grid**2)
z = np.sin(r) / r  # sinc 函数

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

fig.update_layout(
    title="Interactive 3D Sinc Function (Try rotating me!)",
    autosize=True,
    width=800,
    height=600,
    margin=dict(l=65, r=50, b=65, t=90),
    scene=dict(xaxis_title="X Axis", yaxis_title="Y Axis", zaxis_title="Z Amplitude"),
)

fig.show()
```

```python
import plotly.express as px

df = px.data.stocks()

fig = px.line(
    df,
    x="date",
    y=df.columns[1:],
    title="Tech Stocks Over Time (Zoom with the slider below!)",
)

fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

### 6. 进阶：魔法命令

Jupyter 还支持一些特殊的“魔法命令”，以 `%` 或 `%%` 开头，可以实现一些高级功能。

*   `%timeit`: 测量代码运行时间。
*   `%%writefile`: 将 Cell 内容写入文件。
*   `%matplotlib inline`: 在 Notebook 内直接显示 Matplotlib 图表。

```python
print("测试生成一百万个随机数的时间：")
%timeit np.random.rand(1000000)
```
