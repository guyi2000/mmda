### Problem 5

```python
a = np.array([2, -1, 0, -1])
f4 = np.array([1, -1, 1, -1]).T
C4 = la.toeplitz(a)
C4
```

```python
np.linalg.solve(C4, f4)
```

```python
np.matmul(np.linalg.pinv(C4), f4)
```

```python
np.linalg.lstsq(C4, f4, rcond=None)[0]
```

### Problem 20

```python
def isToeplitz(A):
    m, n = A.shape
    # If symmetric
    if m != n or A.T.all() != A.all():
        return False
    # If Toeplitz
    for k in range(m - 1):
        for i in range(m - 2):
            j = -i + k
            if A[i][j] != A[i + 1][j + 1]:
                return False
    return True
```
